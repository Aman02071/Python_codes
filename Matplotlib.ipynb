Data Visualization

```python
import matplotlib.pyplot as plt
```

```python
name =['ram','shyam','suresh','mahesh','virat']
marks =[45,56,90,89,99]
plt.figure(figsize=(8,10))
plt.plot(name,marks,markerfacecolor="orange",marker="D",markeredgecolor="black",
linestyle ="dashed")
plt.xlabel('name')
plt.ylabel('marks')
plt.title('marksheet')
plt.show()
```

```python
Dish = ["egg","chicken","chappati","paneer","gulab_jamun","rajma"]
order_no =[5,10,23,56,67,34]
plt.figure(figsize=(8,10))
plt.plot(Dish,order_no,markerfacecolor="brown",linestyle="dashed",marker="*",
         markeredgecolor="black")
plt.xlabel("dish")
plt.ylabel("order_no")
plt.title("order_details")
plt.show()
```

```python
college =["iit Delhi","vit","bit pilani","Nimms Mumbai","ISC bangalore"]
year =[1200,990,2005,2107,2014]
plt.figure(figsize=(5,5))
plt.bar(college,year,color=["red","blue","brown","green","pink"],width=0.5)
plt.xlabel("college")
plt.ylabel("year")
plt.title("college_year")
plt.show()
```

```python
vehicle_company = ["Hyundai","Suzuki","TATA","Mahindra","kia"]
company_rev = [14,72,67,56,67]
plt.pie(company_rev,labels=vehicle_company,explode=[0.1,0,0,0,0.2],shadow=True,
        colors=['red','green','brown','pink','grey'], autopct = '%1.1f%%')

plt.title("vehicle_company")
plt.show()
```

```python
E_commerce =["Amazon","flipkart","snapdeal","shopclues","myntra","swiggy"]
vistor=[50,35,15,7,28,16]
plt.figure(figsize=(6,8))
plt.pie (vistor,labels=E_commerce,explode=[0,0.2,0,0,0.1,0.1],shadow=True ,colors
        =["grey","yellow","pink","red","blue","brown"],autopct='%1.1f%%')
plt.title("E_commerce")
plt.show()
```

```python
x= [2,4,6,8,10,12,14,16]
y =[10,20,10,40,50,24,35,62]
plt.figure(figsize=(6,7))
plt.grid()
plt.scatter(x,y,color ="blue",marker="*",s =100,alpha =0.7)
plt.title("criss-cross")
plt.xlabel("vizion")
plt.ylabel("crizon")
plt.show()
```
